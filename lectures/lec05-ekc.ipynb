## 2. Columns of Tables are Arrays

Back to the NBA — this time paying attention to what each operation *returns*.

### 📋 Board Reference

| Code | What it does |
|---|---|
| `t.num_rows`, `t.num_columns` | table dimensions (no parentheses!) |
| `t.labels` | the column names |
| `t.relabeled('old', 'new')` | new table with a renamed column |
| `t.select('A')` | one-column **Table** |
| `t.column('A')` | that column as an **array** |
| `t.take(0)` | new table containing just row 0 |

````python
nba = Table.read_table('nba_salaries.csv').where('season', 2020).drop('season')
nba.show(5)
````

````python
nba.num_rows
````

````python
nba.num_columns
````

````python
nba.labels
````

````python
warriors = nba.where('team', 'Golden State Warriors')
warriors.show(5)
````

````python
warriors.relabeled('salary', '$')
````

````python
warriors
````

````python
warriors = nba.where('team', 'Golden State Warriors').drop('team').relabeled('salary', '$')
warriors.show(3)
````

````python
warriors.select('$')
````

````python
type(warriors.select('$'))
````

````python
# This will produce an error
np.average(warriors.select('$'))
````

````python
warriors.column('$')
````

````python
type(warriors.column('$'))
````

````python
np.average(warriors.column('$'))
````

> **Worth repeating:** `.select` → Table, `.column` → array. The error two cells up is what happens when you hand `np.average` a Table. 😬

````python
# another way
warriors.column(3)
````

````python
suns = nba.where('team', 'Phoenix Suns').drop('team').relabeled('salary', '$')
````

````python
np.average(warriors.column('$')) - np.average(suns.column('$'))
````

````python
np.sum(warriors.column('$')) - np.sum(suns.column('$'))
````

## 4. Ways to Create a Table

Two ways: build from arrays with `Table().with_columns(...)`, or read a file with `Table.read_table(...)`. From scratch first — the streets south of campus:

````python
streets = make_array('14th', '13th', '12th', '11th')
streets
````

````python
Table()
````

````python
southside = Table().with_column('Streets', streets)
southside
````

````python
southside.with_column('Blocks from campus', np.arange(4))
````

````python
southside
````

````python
southside = southside.with_column('Blocks from campus', np.arange(4))
southside
````

````python
Table().with_columns(
    'Streets', streets,
    'Blocks from campus', np.arange(4)
)
````

### Reading a Table from a File

> **About this dataset:** W. E. B. Du Bois assembled these income and expenditure data on Black families in Atlanta for the 1900 Paris Exposition, alongside hand-drawn charts that remain some of the most celebrated visualizations ever made. (KL deck, slide 13.) Each row is an income bracket; the expense columns are *proportions* of income.

````python
du_bois = Table.read_table('du_bois.csv')
du_bois
````

````python
du_bois.column('ACTUAL AVERAGE')
````

````python
du_bois.column('FOOD')
````

````python
du_bois.column('ACTUAL AVERAGE') * du_bois.column('FOOD')
````

````python
food_dollars = du_bois.column('ACTUAL AVERAGE') * du_bois.column('FOOD')
du_bois.with_column('Food $', food_dollars)
````

````python
du_bois = du_bois.with_column('Food $', food_dollars)
du_bois
````

````python
np.max(du_bois.column('Food $'))
````

````python
du_bois.with_column(
    'Fraction of well-to-do food $', 
    du_bois.column('Food $') / np.max(du_bois.column('Food $')))
````
