### A few more things we saw in lab

`where` can do much more than exact matches — pass it a **predicate** from the `are` family.

### 📋 Board Reference — `where` predicates

| Code | Rows kept |
|---|---|
| `t.where('A', value)` | A equals value |
| `t.where('A', are.equal_to(x))` | same thing, spelled out |
| `t.where('A', are.above(x))` | A > x |
| `t.where('A', are.above_or_equal_to(x))` | A ≥ x |
| `t.where('A', are.between(x, y))` | x ≤ A < y (y not included!) |
| `t.where('A', are.between_or_equal_to(x, y))` | x ≤ A ≤ y |

````python
# where operation can take on different predicates
nba.where('salary',are.above(1000000))
````

````python
nba.where('salary',are.above_or_equal_to(1000000))
````

````python
nba_2000_2001 = nba.where('season',are.between(2000,2001))
nba_2000_2001
````

````python
nba_2000_2001.sort('season',descending=True)
````

````python
nba.where('season',are.between_or_equal_to(2000,2001)).sort('season',descending=True)
````

## 5. Arrays

An **array** is a sequence of values, and it's how tables store columns. The superpower: arithmetic happens to **every element at once** — no loops needed.

### 📋 Board Reference

| Code | What it does |
|---|---|
| `make_array(1, 2, 3)` | build an array |
| `arr * 2`, `arr + 1` | elementwise arithmetic |
| `arr1 + arr2` | add matching elements (lengths must match!) |
| `arr.item(0)` | first element — counting starts at 0! |
| `len(arr)` | number of elements |
| `sum(arr)`, `np.average(arr)` | total and mean |

````python
first_four = make_array(1, 2, 3, 4)
````

## 6. Columns of Tables are Arrays

Here's where arrays meet tables — and the distinction that matters most this week.

````python
nba = Table.read_table('nba_salaries.csv').where('season', 2020).drop('season')
nba.show(5)
````

````python
warriors = nba.where('team', 'Golden State Warriors')
warriors.show(5)
````

````python
warriors.select('salary')
````

````python
type(warriors.select('salary'))
````

````python
warriors.column('salary')
````

````python
type(warriors.column('salary'))
````

> **The distinction that matters:** `.select('salary')` gives you a **Table**; `.column('salary')` gives you an **array**. Math like `np.average` needs the *array*. You'll use `.column` constantly on Homework 2. 💡

````python
np.average(warriors.column('salary'))
````

````python
suns = nba.where('team', 'Phoenix Suns')
````

````python
np.average(warriors.column('salary')) - np.average(suns.column('salary'))
````

````python
np.sum(warriors.column('salary')) - np.sum(suns.column('salary'))
````

````python
warriors.num_rows
````

````python
suns.num_rows
````
