## 5. Back to the Dice Game — and Growing Arrays

Instead of making up rolls, let the computer roll. Then, to keep score across many games, we need an array that **grows** — that's `np.append`.

````python
die_faces = np.arange(1, 7)
````

````python
np.random.choice(die_faces)
````

````python
def simulate_one_round():
    my_roll = np.random.choice(die_faces)
    your_roll = np.random.choice(die_faces)
    return one_round(my_roll, your_roll)
````

````python
simulate_one_round()
````

## 6. `for` Statements

Running that cell by hand 10,000 times would take a while. 😅 A `for` loop repeats the body once per element — this is the single most important pattern of the next ten lectures (KL deck, slides 14–16):

```python
for i in np.arange(repetitions):
    ...simulate one outcome, append it...
```

````python
for pet in make_array('cat', 'dog', 'rabbit'):
    print('I love my ' + pet)
````

````python
pet = make_array('cat', 'dog', 'rabbit').item(0)
print('I love my ' + pet)

pet = make_array('cat', 'dog', 'rabbit').item(1)
print('I love my ' + pet)

pet = make_array('cat', 'dog', 'rabbit').item(2)
print('I love my ' + pet)
````

````python
game_outcomes = make_array()

for i in np.arange(5):
    game_outcomes = np.append(game_outcomes, simulate_one_round())
    
game_outcomes
````

````python
game_outcomes = make_array()

for i in np.arange(10000):
    game_outcomes = np.append(game_outcomes, simulate_one_round())
    
game_outcomes
````

````python
len(game_outcomes)
````

````python
results = Table().with_column('My winnings', game_outcomes)
results
````

````python
results.group('My winnings').barh('My winnings')
````

> **Read that bar chart:** in 10,000 simulated games the three outcomes came out nearly even — exactly what the symmetry of the game predicts. Simulation just *verified* a theory. 🧪

And here's the full simulation recipe — **define one outcome, repeat, collect, visualize.** You'll use this exact skeleton in Lab 5 and Homework 6:

````python
# Simulate one outcome

def num_heads():
    return sum(np.random.choice(coin, 100) == 'heads')
````

````python
# Decide how many times you want to repeat the experiment

repetitions = 10000
````

````python
# Simulate that many outcomes

outcomes = make_array()

for i in np.arange(repetitions):
    outcomes = np.append(outcomes, num_heads())
````

````python
heads = Table().with_column('Heads', outcomes)
heads.hist(bins = np.arange(29.5, 70.6))
````

### More Conditional Statements

One more `if` in the wild: classifying every bike trip as round trip or one way, with `apply` + `pivot` to finish.

````python
trip = Table().read_table('trip.csv')
trip.show(3)
````

````python
def trip_kind(start, end):
    if start == end:
        return 'round trip'
    else:
        return 'one way'

kinds = trip.with_column('Trip Kind', trip.apply(trip_kind, 'Start Station', 'End Station'))
kinds.show(3)
````

````python
kinds.where('Duration', are.below(600)).pivot('Trip Kind', 'Start Station')
````
