## 4. Tables 🍦

Enough warm-up — meet the star of the course. A **Table** has labeled columns and one row per record. Our first one is small and delicious.

### 📋 Board Reference

| Code | What it does |
|---|---|
| `Table.read_table('file.csv')` | load a table from a file |
| `t.show(3)` | display the first 3 rows |
| `t.select('A', 'B')` | new table with only those columns |
| `t.drop('A')` | new table without that column |
| `t.where('A', value)` | only rows where column A has that value |
| `t.sort('A')` | sort rows by column A, smallest first |
| `t.sort('A', descending=True)` | sort largest first |

````python
from datascience import *
````

````python
cones = Table.read_table('cones.csv')
cones
````

````python
cones.show(3)
````

````python
cones.show()
````

````python
cones.select('Flavor')
````

````python
cones.select('Flavor', 'Price')
````

````python
# The line below causes an error because Flavor is not a name.
# cones.select(Flavor, 'Price')
````

````python
cones.drop('Price')
````

````python
cones
````

````python
cones_without_price = cones.drop('Price')
cones_without_price
````

> **Table methods return a *new* table** — `cones` itself never changed. To keep a result, give it a name: `cones_without_price = cones.drop('Price')`.

````python
cones.where('Flavor', 'chocolate')
````

````python
cones.where('Price',5.25)
````

````python
cones.sort('Price')
````

````python
cones.sort('Price', descending=True)
````

````python
cones.sort('Flavor', descending=True)
````
