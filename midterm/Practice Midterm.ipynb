## Correct Option a

```python
def count_arr_occurences(arr, value):
    count = 0
    for x in arr:
        if x == value:
            count = count + 1
    return count
```

```python
count_arr_occurences(make_array(1,3,5,7,7), 7)
```

# Question 5

```python
restaurants = Table().with_columns('Restaurant', make_array("Renata's", "Mazzoti's", "Bayside", "Pho", "Sushi Spot"))
restaurants = restaurants.with_columns('Type', make_array('Cafe', 'Italina', 'Cafe', 'Vietnamese', 'Sushi'))
restaurants = restaurants.with_columns('Distance from Campus', make_array(.5,.7,2,.3,.5))
restaurants
```

```python
transport = restaurants = Table().with_columns('Restaurant', make_array("Renata's", "Bayside", "Sushi Tao", "Mazotti's", "Mazzotti's"))
transport = transport.with_column("Transportation", make_array('Walk', "Drive", "Bike", "Bike", "Drive"))
transport = transport.with_column("Time", make_array(14, 7, 4, 6, 4))
transport
```

```python
def first(some_array):
    return some_array.item(0)
```

## a 
table with fastest transportation method to get to each restaurant

```python
transport.sort('Time').select("Restaurant", "Transportation").group("Restaurant", first)
```

## b 
produce table with fastest time for any type of food

```python
restaurants = Table().with_columns('Restaurant', make_array("Renata's", "Mazzoti's", "Bayside", "Pho", "Sushi Spot"))
restaurants = restaurants.with_columns('Type', make_array('Cafe', 'Italina', 'Cafe', 'Vietnamese', 'Sushi'))
restaurants = restaurants.with_columns('Distance from Campus', make_array(.5,.7,2,.3,.5))
restaurants
```

```python
restaurants.join("Restaurant", transport, "Restaurant").select("Type", "Time").group("Type", first)
```

## c
produce table with columns corresponding to each unique transportation mode and rows corresponding to each unique restaurant type, cells containing minimum travel time.

```python
restaurants.join("Restaurant", transport, "Restaurant").pivot("Transportation", "Type", "Time", min)
```

## 7 
write function `get_largest` which takes 3 numbers and returns the largest of the three (hint: what if 2 of the numbers are equal?)

```python
def get_largest(a, b, c):
    if a >= b and a >= c:
        largest = a
    elif b >= a and b >= c:
        largest = b
    else:
        largest = c
    return largest    
```

```python
get_largest(1,2,3)
```

```python
get_largest(5,5,7)
```

```python
get_largest(5,5,4)
```
