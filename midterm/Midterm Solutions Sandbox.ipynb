```python
def chance_of_marbles(color1, color2):
    total_marbles = sum(colors.column('number'))
    num_color1 = colors.where('color', color1).column('number').item(0)
    num_color2 = colors.where('color', color2).column('number').item(0)
    if color1 == color2:
        chance = (num_color1 * (num_color1 - 1)) / (total_marbles * (total_marbles - 1))
    else:
        chance = (num_color1 * num_color2) / (total_marbles * (total_marbles - 1))
    return chance
```
