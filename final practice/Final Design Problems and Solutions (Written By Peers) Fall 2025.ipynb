# Problem 2
Which expression correctly computes the correlation between two columns x and y in a table t?

# Solution 2 

```python
x_arr = t.column(x)
y_arr = t.column(y)
x_standard = (x_arr - np.average(x_arr)) / np.std(x_arr)
y_standard = (y_arr - np.average(y_arr)) / np.std(y_arr)

1 = np.average(x_standard * y_standard)

2 = (x_arr.std() * y_arr.std()) / np.average(x_arr * y_arr)

3 = np.average((x_arr - x_arr.mean()) / np.std(y_arr))
```

# Problem 4
Create a function that takes in a table and returns a bootstrapped sample.

# Solution 4

```python
def bootstrap(tbl):
    n = tbl.num_rows
    bs_tbl = tbl.sample(n, with_replacement=True)
    return bs_tbl
```

```python
def one_test_stat():
    shuffled = pizza.sample(with_replacement=False).column('City')
    shuffled_table = pizza.with_column('City_Shuffled', shuffled)
    mean_A = shuffled_table.where('City_Shuffled', 'CityA').column('Cheese_Ratio').mean()
    mean_B = shuffled_table.where('City_Shuffled', 'CityB').column('Cheese_Ratio').mean()
    return mean_A - mean_B
```

# Problem 10
Find the Euclidean Distance between the first two rows of the dataset provided below. Use the "Critical Reading", "Math", and "Writing" columns. You can use the provided function `row_to_array` if you would like.  Use the file `sat2014.csv`.

```python
sat = Table.read_table('sat2014.csv')
sat.show(3)
def row_to_array(table, row):
    '''take a table and desired row number and outputs the chosen row as an array of values.'''
    row_list = table.row(row)
    return np.array(row_list)
```

# Solution 10

```python
sat_only_scores = sat.select("Critical Reading", "Math", "Writing")

#Using the function
dist_fn = (sum((row_to_array(sat_only_scores, 0) - row_to_array(sat_only_scores, 1))**2))**0.5

#Print Results
print("The euclidean distance between the first two rows of the dataset is", dist_fn.round(3))
```

```python
sat_only_scores = sat.select("Critical Reading", "Math", "Writing")

# Without using the function
row0 = np.array(sat_only_scores.row(0))
row1 = np.array(sat_only_scores.row(1))

dist_no_fn = (sum((row0 - row1)**2))**0.5

#Print Results
print("The euclidean distance between the first two rows of the dataset is", dist_no_fn.round(3))
```
